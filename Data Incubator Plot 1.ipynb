```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("checkouts-by-title_trueSample.csv")

trainDF = pd.DataFrame()
trainDF = df.loc[df.loc[:,'MaterialType']=="BOOK",['Checkouts','Subjects']]
trainDF = trainDF.dropna(axis=0)

keyDict = {"mystery":0, "science fiction":0, "adventure":0, "fantasy":0, "tragedy":0, "horror":0, "romance":0}
for row in trainDF.iterrows():
    for g in keyDict.keys():
        if(g in row[1]['Subjects'].lower()):
           keyDict[g] += row[1]['Checkouts']

plt.bar(keyDict.keys(), keyDict.values());
```

We see different genres having vastly different total checkouts, suggesting that it could be a good predictor.
