```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv( "checkouts-by-title_trueSample.csv")

trainDF = pd.DataFrame()
trainDF = df.loc[df.loc[:,'MaterialType']=="BOOK",['Checkouts', 'CheckoutYear','Subjects']]
trainDF = trainDF.dropna(axis=0)

trainDF = trainDF[trainDF["Subjects"].str.contains("Python Computer program language")]

plot = trainDF.groupby(['CheckoutYear'])['Checkouts'].sum().plot()
```

We see the growing popularity of python as a programming language coinciding with the rise of data science as a profession
